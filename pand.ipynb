```python
import numpy as np
import statsmodels.api as sm
gaussian_rv = np.random.normal(loc=0,scale=1,size=1000) # generate gaussian rand variable
skew_right=np.concatenate((gaussian_rv[gaussian_rv > 0] * 2.5, gaussian_rv)) # create skewed ryt Ds
plt.hist(skew_right,bins=100,alpha=0.5,label="Skewed Right") # plot histogram of the skewed Ryt Ds

sm.qqplot(skew_right) # plot the qq plot of the skewed ryt Ds
plt.legend()
plt.show()

skew_left=np.concatenate((gaussian_rv[gaussian_rv < 0] * 2.5, gaussian_rv)) # create skewed left Ds
plt.hist(skew_left,bins=100,alpha=0.5,label="Skewed left")

sm.qqplot(skew_left) # plot the qq plot of the skewed left Ds
plt.legend()
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
from scipy.stats import norm
from statsmodels.graphics.gofplots import qqplot

gaussian_rv=np.random.normal(loc=0,scale=1,size=1000)
fat_tails=np.concatenate((gaussian_rv  * 2.5, gaussian_rv))
plt.hist(fat_tails,bins=100,alpha=0.5,label="fat_tails")

qqplot(fat_tails,line="45")
plt.legend()
plt.show()

thin_tails=np.random.normal(loc=0,scale=0.7,size=1000)
plt.hist(thin_tails,bins=100,alpha=0.5,label="thin_tails")

qqplot(thin_tails,line="45")
plt.legend()
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats
import matplotlib.pyplot as plt
x=np.random.uniform(low=0,high=1,size=1000) #generate set of random data
```

```python
plt.hist(x)
```

```python
params = stats.uniform.fit(x)
dist=stats.uniform(loc=params[0],scale=params[1])
```

```python
fig=sm.qqplot(x,dist=dist,line="45") #using uniform distribution
plt.title("QQ plot of uniform Distribution with uniform fit")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.show()
```

```python
num_samples=100000
samples_size=300
distribution_range=(0,1)
#generate samples from a uniform distribution
samples=np.random.uniform(distribution_range[0],distribution_range[1],(num_samples,samples_size))
sample_means=np.mean(samples,axis=1) # calculate sam mean
plt.hist(sample_means,bins=30,density=True,edgecolor="white",color="Black")
plt.title("Histogram of sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
num =1000 #set parameter
sample_size=50
lamda_param=2

samples=np.random.exponential(scale=1/lamda_param, size=(num,sample_size)) #  generate sample from exponential distribution

sample_means=np.mean(samples,axis=1) # calculate sample means

plt.hist(sample_means,bins=30,density=True,edgecolor="black",color="maroon") #plot histogram
plt.title("Histogram of sample Means (Exponential Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
num=1000
sample=50

poisson_lamda=5

gamma_shape=2
gamma_scale=1

binomial_n=10
binomial_p=0.5

poisson_sample=np.random.poisson(lam=poisson_lamda,size=(num,sample))
gamma_sample=np.random.gamma(shape=gamma_shape,scale=gamma_scale,size=(num,sample))
binomial_sample=np.random.binomial(n=binomial_n,p=binomial_p,size=(num,sample))

poisson_means=np.mean(poisson_sample,axis=1)
gamma_means=np.mean(gamma_sample,axis=1)
binomial_means=np.mean(binomial_sample,axis=1)

fig,axs=plt.subplots(3,1,figsize=(8,12))

axs[0].hist(poisson_means,bins=30,density=True,edgecolor="black",color="white")
axs[0].set_title("Histogram of sample Means (Poisson_Distribution)")
axs[0].set_xlabel("Sample Mean")
axs[0].set_ylabel("Density")


axs[1].hist(gamma_means,bins=30,density=True,edgecolor="black",color="blue")
axs[1].set_title("Histogram of sample Means (Gamma_Distribution)")
axs[1].set_xlabel("Sample Mean")
axs[1].set_ylabel("Density")


axs[2].hist(binomial_means, bins=30, density=True, edgecolor="blue", color="cyan")
axs[2].set_title("Histogram of sample Means (Binomial_Distribution)")
axs[2].set_xlabel("Sample Mean")
axs[2].set_ylabel("Density")
     

fig.tight_layout()
plt.show()
```
